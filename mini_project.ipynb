```python
menu = [
    {"name": "아이스아메리카노", "price": 3000},
    {"name": "아메리카노", "price": 2500},
    {"name": "아이스카페라떼", "price": 4000},
    {"name": "카페라떼", "price": 3500},
    {"name": "샷추가 1회", "price": 500},
    {"name": "휘핑 1회", "price": 500},
    {"name": "자바칩 추가", "price": 500},
    {"name": "아이스캬라멜마끼아또", "price": 4500},
    {"name": "캬라멜마끼아또", "price": 4000},
    {"name": "에스프레소", "price": 1500},
    {"name": "소금빵", "price": 3500}
]

#초기 주문 음성인식
uploaded = files.upload()
for file_name in uploaded.keys():
    with open(file_name, "rb") as audio_file:
        transcription_response = openai.Audio.transcribe("whisper-1", audio_file)
        user_order_text = transcription_response["text"]
        print("주문 텍스트 변환:", user_order_text)



#카페 메뉴 정보
menu_info = (
  "다음은 카페 메뉴와 가격입니다:\n"
  "아이스아메리카노 - 3,000원\n"
  "아메리카노 - 2,500원\n"
  "아이스카페라떼 - 4,000원\n"
  "카페라떼 - 3,500원\n"
  "샷추가 1회 - 500원\n"
  "휘핑 1회 - 500원\n"
  "자바칩 추가 - 500원\n"
  "아이스캬라멜마끼아또 - 4,500원\n"
  "캬라멜마끼아또 - 4,000원\n"
  "에스프레소 - 1,500원\n"
  "소금빵 - 3,500원\n\n"
  "고객의 주문을 바탕으로 총액을 계산해주세요."
)

#카페 메뉴를 바탕으로 주문을 받는 gpt
def process_order(order_text):
    messages = [
        {"role": "system", "content": "당신은 카페에서 주문을 받는 친절한 어시스턴트입니다. 고객의 주문을 이해하고, 세부 사항을 확인하며, 다음 메뉴와 가격을 참고하여 총액을 계산하세요:\n\n" + menu_info},
        {"role": "user", "content": order_text}
    ]

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    return response['choices'][0]['message']['content']

#결제 방법 처리
def ask_payment_method():
    messages = [
        {"role": "system", "content": "당신은 카페에서 친절하게 결제를 도와주는 어시스턴트입니다."},
        {"role": "user", "content": "고객에게 결제 방법을 물어보세요. '카드로 결제하시겠습니까? 현금으로 결제하시겠습니까?'."}
    ]

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    return response['choices'][0]['message']['content']

#결제 방법 음성인식 처리
def process_payment():
    payment_question = ask_payment_method()
    print("결제 방법 질문:", payment_question)

    uploaded = files.upload()
    for file_name in uploaded.keys():
        with open(file_name, "rb") as audio_file:
            transcription_response = openai.Audio.transcribe("whisper-1", audio_file)
            payment_method_text = transcription_response["text"]
            print("결제 방법 변환:", payment_method_text)

    #카드 결제
    if "카드" in payment_method_text:
        print("카드 결제가 진행됩니다.")
        print("결제가 완료되었습니다. 감사합니다!")
    #현금 결제
    elif "현금" in payment_method_text:
        print("현금 결제가 진행됩니다.")
        print("결제가 완료되었습니다. 감사합니다!")
    else:
        print("잘못된 결제 방법입니다. 다시 시도해주세요.")

order_response = process_order(user_order_text)
print("주문 처리 결과:", order_response)


# 결제 진행
process_payment()
```
